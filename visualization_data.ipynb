```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('car-data-handle.csv')
```

```python
data.head()
```

```python
data = data.drop(columns=['Unnamed: 0'])
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

columns = ['brand', 'name', 'model', 'fuel', 'origin', 'gear'] 

for column in columns:
  data[column] = le.fit_transform(data[column])
```

```python
data.head()
```

## Tương quan dữ liêu


```python
data_corr = data.select_dtypes(include='number').corr()
```

```python
data_corr
```

```python
ones_corr = np.ones_like(data_corr, dtype=bool)
ones_corr
```

```python
mask = np.triu(ones_corr)
mask
```

```python
fig, ax = plt.subplots(figsize=(8,6))
mask_rmk = mask[1:, :-1]
data_corr_rmk = data_corr.iloc[1: , :-1]

cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
sns.heatmap(data=data_corr_rmk, 
            mask=mask_rmk, 
            annot=True, 
            fmt=".2f", 
            cmap=cmap,
            vmin=-1,
            vmax=1,
            linecolor="White",
            linewidth="0.5")

yticks = [i.upper() for i in data_corr_rmk.index]
xticks = [i.upper() for i in data_corr_rmk.columns]

ax.set_yticklabels(yticks, rotation=0);
ax.set_xticklabels(xticks, rotation=90);
title = "Correlation Matrix";
ax.set_title(title, loc='left');
```
