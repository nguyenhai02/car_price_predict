# DATA MANIPULATION

```python
df = df.rename(columns={'Hãng xe': 'brand', 
                        'Tên xe': 'name', 
                        'Giá (VNĐ)' : 'price',
                        'Năm sản xuất': 'year',
                        'Dòng xe' : 'model',
                        'Số chỗ ngồi' : 'seat',
                        'Nguyên liệu' : 'fuel',
                        'Dung tích (L)' : 'capacity',
                        'Xuất xứ' : 'origin',
                        'Hộp số' : 'gear'})
```

```python
df.head()
```

```python
df.price.dtype #kiểu hiện tại của price là Object
```

```python
df.price = df.price.apply(lambda x : int(x.replace(',', ''))) #chuyển kiểu dữ liệu của price sang int
```

```python
df.price.dtype
```

```python
df.info()
```

```python
df.head()
```

```python
df.name = df.brand + df.name # Hoàn chỉnh tên của xe
```

```python
df.head()
```

```python
price = df.pop('price')
df.insert(len(df.columns), 'price', price)
```
